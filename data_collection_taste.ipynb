```python
import boto3
from boto3 import session
import pandas as pd
```

```python
#making session object
session = boto3.session.Session(region_name='us-west-2')
```

```python
#making s3 client
s3client = session.client('s3', config = boto3.session.Config(signature_version='s3v4'))
```

```python
# creating response
response = s3client.get_object(Bucket='etongue', Key='E-Tongue data/beer11.txt')
```

```python
response
```

after calling the client we generate a resposne object. response['Body'] is the actual file to access

'HTTPStatusCode' tells us if the response request is successful or not. If it not 200, it has failed!

```python
response['Body']
```

```python
contents = response['Body'].read()
```

```python
contents
```

```python
decoded_contents = contents.decode("utf-8")
decoded_contents
```

```python
bucket='etongue'
result = s3client.list_objects(Bucket = bucket, Prefix='E-Tongue data')
```

```python
result
```

    data = s3client.get_object(Bucket=bucket, Key=o.get('Key'))
